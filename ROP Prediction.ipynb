```python
from sklearn import preprocessing

ohe = preprocessing.OneHotEncoder()
ohe_up5 = pd.DataFrame(ohe.fit_transform(data[['Formation']]).toarray())
ohe_up6 = pd.DataFrame(ohe.fit_transform(data1[['FORMATION']]).toarray())
```
